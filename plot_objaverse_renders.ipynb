# Plotting Rendered Objaverse Images with Matplotlib

```python
# get the first image of each object which has been rendered from the front
# The images are strored like views/some_uid/rgba/000.png - where some_uid is the uid of the object
# and 000.png is the image number

# get all subdirectories in the views directory
import os
views_dir = "views"
view_dirs = os.listdir(views_dir)

# get the first image of each object
all_images = []
for view_dir in view_dirs:
    view_dir_path = os.path.join(views_dir, view_dir)
    image_folder_path = os.path.join(view_dir_path, "rgba")
    image_files = os.listdir(image_folder_path)
    image_files.sort()
    first_image = image_files[0]
    first_image_path = os.path.join(image_folder_path, first_image)
    all_images.append(first_image_path)
    second_image = image_files[10]
    second_image_path = os.path.join(image_folder_path, second_image)
    all_images.append(second_image_path)
    third_image = image_files[21]
    third_image_path = os.path.join(image_folder_path, third_image)
    all_images.append(third_image_path)
    fourth_image = image_files[28]
    fourth_image_path = os.path.join(image_folder_path, fourth_image)
    all_images.append(fourth_image_path)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


def display_images(image_paths, columns=10, figsize=(10, 10)):
    """
    Displays images in a grid.
    Args:
    image_paths (list of str): List of image file paths
    columns (int): Number of columns in the image grid
    figsize (tuple): Dimension of the figure (width, height)
    """
    # Load images
    images = [mpimg.imread(path) for path in image_paths]

    # Calculate number of rows needed
    rows = (len(images) + columns - 1) // columns

    # Create a figure with subplots
    fig, axs = plt.subplots(rows, columns, figsize=figsize)
    axs = axs.flatten()  # Flatten to 1D array

    for img, ax in zip(images, axs):
        ax.imshow(img)
        ax.axis('off')  # Hide axes

    # # Hide unused subplots
    # for ax in axs[len(images):]:
    #     ax.axis('off')

    plt.show()
```

```python
# Display the first images of the car objects
sub_image_list = all_images[:8]
figsize = (60, 16)
display_images(sub_image_list, columns=8, figsize=figsize)
sub_image_list = all_images[8:16]
display_images(sub_image_list, columns=8, figsize=figsize)
```

```python
# go throuh all images in steps of 8 and call the display_images function
test_sub_image_list = all_images[:100]
step_size = 8
for i in range(0, len(test_sub_image_list), step_size):
    sub_image_list = test_sub_image_list[i:i+step_size]
    display_images(sub_image_list, columns=step_size, figsize=(60, 16))
```
